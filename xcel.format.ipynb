```python
import pandas as pd
from google.colab import files
import io
from openpyxl import load_workbook
from openpyxl.styles import Font, Border, Side, PatternFill
from openpyxl.utils.dataframe import dataframe_to_rows

# Step 1: Upload the Excel file
uploaded = files.upload()

# Assuming only one file is uploaded, we take the first file name
file_name = list(uploaded.keys())[0]

# Step 2: Load the uploaded Excel file into a DataFrame
df = pd.read_excel(io.BytesIO(uploaded[file_name]))

# Drop the first column
df.drop(df.columns[0], axis=1, inplace=True)

# Sort the DataFrame by the 'TEAM NAME' column alphabetically
df.columns = df.columns.str.strip()  # Remove any extra whitespace from column names
df_sorted = df.sort_values(by='TEAM NAME')

# Step 3: Save the edited DataFrame to a new Excel file
output_file_name = 'ANN_Lab_teams_edited.xlsx'
df_sorted.to_excel(output_file_name, index=False)

# Step 4: Format the Excel file
# Load the workbook and select the active sheet
wb = load_workbook(output_file_name)
ws = wb.active

# Define some styles for formatting
header_font = Font(bold=True, color="FFFFFF")
header_fill = PatternFill(start_color="4F81BD", end_color="4F81BD", fill_type="solid")
border_style = Border(left=Side(style='thin'), right=Side(style='thin'), top=Side(style='thin'), bottom=Side(style='thin'))

# Apply styles to header row
for cell in ws[1]:
    cell.font = header_font
    cell.fill = header_fill
    cell.border = border_style

# Apply border to all cells
for row in ws.iter_rows(min_row=1, max_row=ws.max_row, min_col=1, max_col=ws.max_column):
    for cell in row:
        cell.border = border_style

# Auto adjust column widths
for col in ws.columns:
    max_length = 0
    column = col[0].column_letter  # Get the column name
    for cell in col:
        try:
            if len(str(cell.value)) > max_length:
                max_length = len(cell.value)
        except:
            pass
    adjusted_width = (max_length + 2)
    ws.column_dimensions[column].width = adjusted_width

# Save the formatted file
formatted_output_file_name = 'ANN_Lab_teams_edited_formatted.xlsx'
wb.save(formatted_output_file_name)

# Step 5: Download the formatted file
files.download(formatted_output_file_name)
```
